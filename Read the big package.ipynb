```python
import os
import pandas as pd

# Correct the path to your directory
directory_path = 'C:/Users/张天羽/Desktop/fut_data_1min_2023'  # Update this to your actual directory path

dataframes = {}

for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory_path, filename)
        # Read the CSV with specified encoding
        df = pd.read_csv(file_path, encoding='gbk')
        # Store the DataFrame in a dictionary
        dataframes[filename] = df
        # Print the filename
        print(f"First 5 rows of {filename}:")
        # Print the first 5 rows of the DataFrame
        print(df.head(), '\n')  # Add an extra newline for better separation between files

# Now you have a dictionary of DataFrames indexed by filename, and you've printed the first 5 rows of each
```
